```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Predictive Modelling for Malaria Prevention - Water (1).csv')
df.head()
```

```python
print(df.columns)
```

```python
df = df.drop([1, 2, 4, 5])
df = df.reset_index(drop=True)
df.head()
```

```python
# checking missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Descriptive statistics
stat = df.describe()
print("Descriptive Statistics:\n", stat)
```

```python
# Convert numeric columns to numeric, replace commas and handle non-numeric values
numeric_columns = ['CIV', 'CIV.1', 'GIN', 'GIN.1', 'LBR', 'LBR.1', 'NGA', 'NGA.1', 'SLE', 'SLE.1']

for column in numeric_columns:
    df[column] = pd.to_numeric(df[column].str.replace(',', ''), errors='coerce')

# Descriptive statistics after converting to numeric
stat_numeric = df[numeric_columns].describe()
print("Descriptive Statistics (Numeric):\n", stat_numeric)

```

```python
import pandas as pd
# columns for improved sanitation facilities
sanitation_columns = ['CIV', 'GIN','LBR', 'NGA', 'SLE']
sanitation_max_values = df[sanitation_columns].max()

# columns for improved water sources
water_columns = ['CIV.1','GIN.1', 'LBR.1', 'NGA.1', 'SLE.1']
water_max_values = df[water_columns].max()

# Country with maximum values for improved sanitation facilities and water sources
max_sanitation_country = sanitation_max_values.idxmax()
max_water_country = water_max_values.idxmax()

print(f"Country with the highest improved sanitation facilities: {max_sanitation_country} ({sanitation_max_values[max_sanitation_country]}%)")
print(f"Country with the highest improved water sources: {max_water_country} ({water_max_values[max_water_country]}%)")
```
