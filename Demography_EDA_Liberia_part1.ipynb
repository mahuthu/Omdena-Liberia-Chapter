```python
df1 = df.loc[np.where(df['Indicator Name'] == "Population, total")]
df1.rename(columns = {'Value':'Total_population'},inplace=True)

df2 = df[['Value','Year']].loc[np.where(df['Indicator Name'] == "Population, female")]
df2.rename(columns = {'Value':'no_of_females'},inplace=True)

df3 = df[['Value','Year']].loc[np.where(df['Indicator Name'] == "Population, male")]
df3.rename(columns = {'Value':'no_of_males'},inplace=True)
```

```python
df5 = df1.merge(df2,on='Year').merge(df3,on='Year')
```

```python
df5['Percentage_of_females'] = (np.divide(df5['no_of_females'],df5['Total_population'])) * 100
```

```python
df5['Percentage_of_males'] = (np.divide(df5['no_of_males'],df5['Total_population'])) * 100
```

**Pearson Correlation**: Also known as linear correlation, Pearson correlation measures the strength and direction of a linear relationship between two continuous variables.
