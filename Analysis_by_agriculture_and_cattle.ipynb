```python
df = pd.read_excel("/content/drive/MyDrive/Malaria_prediction/Prevalence_of_malaria_in_children.xlsx",header=3)
```

```python
df.head()
```

```python
df.info()
```

```python
df.fillna(value='null',inplace=True)
```

```python
import re
```

```python
# Define a function to extract numeric values from the strings
def extract_numeric_value(text):
    match = re.search(r'(\d+\.\d+)',text)
    if match:
        return float(match.group(0))
    else:
        return None
```

```python
df['Malaria prevalence according to RDT'] = df['Malaria prevalence according to RDT'].apply(extract_numeric_value)
df['Malaria prevalence according to microscopy'] = df['Malaria prevalence according to microscopy'].apply(extract_numeric_value)
```

```python
df.head()
```

```python
df_2022_MIS = df.loc[np.where(df['Survey']=='2022 MIS')]
```

```python
df_2022_MIS['Characteristic']
```

```python
ax = df_2022_MIS[19:26].groupby('Characteristic').sum().plot.barh()
ax.set_title("Malaria prevalence in children by Age-Group")
ax.set(xlabel = 'Percentage',ylabel = 'Age Group')
ax.legend(loc=2, bbox_to_anchor=(1.0,1.0));
```

```python
ax = df_2022_MIS[5:9].groupby('Characteristic').sum().plot.barh()
ax.set_title("Malaria prevalence by Education level")
ax.set(xlabel = 'Percentage',ylabel = 'Education level')
ax.legend(loc=2, bbox_to_anchor=(1.0,1.0));
```

```python
ax = df_2022_MIS[1:3].groupby('Characteristic').sum().plot.barh()
ax.set_title("Malaria prevalence in children by Gender")
ax.set(xlabel = 'Percentage',ylabel = 'Gender')
ax.legend(loc=2, bbox_to_anchor=(1.0,1.0));
```

```python
ax = df_2022_MIS[3:5].groupby('Characteristic').sum().plot.barh()
ax.set_title("Malaria prevalence in children by Residence")
ax.set(xlabel = 'Percentage',ylabel = 'Residence')
ax.legend(loc=2, bbox_to_anchor=(1.0,1.0));
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.kdeplot(merged_df['Malaria prevalence according to RDT'], fill=True)
plt.title('Kernel Density Estimation Plot')
plt.xlabel('Percentage')
plt.ylabel('Density')
plt.show()
```

```python
sns.kdeplot(merged_df['Households owning farm animals'], fill=True)
plt.title('Kernel Density Estimation Plot')
plt.xlabel('Percentage')
plt.ylabel('Density')
plt.show()
```

```python
corr_matrix_pearson = merged_df.corr('spearman')
# Plot correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix_pearson, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Plot - Spearman')
plt.show()
```

```python
corr_matrix_pearson = merged_df.corr('kendall')
# Plot correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix_pearson, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Plot - Kendall')
plt.show()
```
